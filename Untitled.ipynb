```python
genres = pd.DataFrame(response.json()['genres'])
genres
```

```python
def merge_genres(genre_ids):
    genre_names = []
    for genre_id in genre_ids:
        genre_name = genres[genres['id'] == genre_id]['name'].values[0]
        genre_names.append(genre_name)
    return genre_names

# Apply the function to merge genre names to the 'genre_ids' column in movies_df
movies['genre_names'] = movies['genre_ids'].apply(merge_genres)
```
